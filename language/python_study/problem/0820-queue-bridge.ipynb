```python
from collections import deque
def solution(bridge_length, weight, truck_weights):
    wait = deque(truck_weights)
    #다리 상태
    remain_len = bridge_length
    remain_weight = weight
    #시간상태
    time = 0
    while wait :
        current = wait.popleft()
        if current <= remain_weight and remain_len :
            time += 1
            remain_weight -= current
            remain_len -= 1
        else : 
            wait.appendleft(current)
            time += bridge_length
            remain_len = bridge_length
            remain_weight = weight
    if not wait :
        time += bridge_length
    return time
```
