# MooTube 문제

## 문제 설명

농부 존은 남는 시간에 **MooTube**라 불리는 동영상 공유 서비스를 만들었다. MooTube에서 농부 존의 소들은 재밌는 동영상들을 서로 공유할 수 있다. 

소들은 MooTube에 1부터 N까지 번호가 붙여진 **N개의 동영상** (1 ≤ N ≤ 5,000)을 이미 올려 놓았다. 하지만, 존은 아직 어떻게 하면 소들이 그들이 좋아할 만한 새 동영상을 찾을 수 있을지 괜찮은 방법을 떠올리지 못했다.

### USADO 측정 시스템

농부 존은 모든 MooTube 동영상에 대해 **"연관 동영상"** 리스트를 만들기로 했다. 이렇게 하면 소들은 지금 보고 있는 동영상과 연관성이 높은 동영상을 추천 받을 수 있을 것이다.

존은 두 동영상이 서로 얼마나 가까운 지를 측정하는 단위인 **"USADO"**를 만들었다. 존은 N-1개의 동영상 쌍을 골라서 직접 두 쌍의 USADO를 계산했다.

### 네트워크 구조

존은 이 동영상들을 네트워크 구조로 바꿔서, 각 동영상을 정점으로 나타내기로 했다. 또 존은 동영상들의 연결 구조를 서로 연결되어 있는 N-1개의 동영상 쌍으로 나타내었다. 

> 💡 **핵심**: 존은 N-1개의 동영상 쌍을 골라서 어떤 동영상에서 다른 동영상으로 가는 경로가 반드시 하나 존재하도록 했다.

**임의의 두 동영상 사이의 USADO**는 그 경로의 모든 연결들의 USADO 중 **최솟값**으로 정의된다.

### 추천 시스템

존은 어떤 주어진 MooTube 동영상에 대해, 값 K를 정해서 그 동영상과 **USADO가 K 이상인 모든 동영상**이 추천되도록 할 것이다. 

하지만 존은 너무 많은 동영상이 추천되면 소들이 일하는 것이 방해될까 봐 걱정하고 있다! 그래서 그는 K를 적절한 값으로 결정하려고 한다.

---

## 입력

**첫 번째 줄**: N과 Q (1 ≤ Q ≤ 5,000)

**다음 N-1개의 줄**: 농부 존이 직접 잰 두 동영상 쌍의 USADO
- 각 줄은 세 정수 `pi, qi, ri` (1 ≤ pi, qi ≤ N, 1 ≤ ri ≤ 1,000,000,000)
- 동영상 pi와 qi가 USADO ri로 서로 연결되어 있음을 의미

**다음 Q개의 줄**: 농부 존의 Q개의 질문
- 각 줄은 두 정수 `ki, vi` (1 ≤ ki ≤ 1,000,000,000, 1 ≤ vi ≤ N)
- K = ki일 때 동영상 vi를 보고 있는 소들에게 몇 개의 동영상이 추천될지 묻는 질문

---

## 출력

Q개의 줄을 출력한다. i번째 줄에는 농부 존의 i번째 질문에 대한 답변이 출력되어야 한다.

---

## 예제

### 입력
```
4 3
1 2 3
2 3 2
2 4 4
1 2
4 1
3 1
```

### 출력
```
3
0
2
```

## 힌트

농부 존은 다음과 같은 USADO 값을 설정했다:
- 1번 동영상 ↔ 2번 동영상: USADO 3
- 2번 동영상 ↔ 3번 동영상: USADO 2  
- 2번 동영상 ↔ 4번 동영상: USADO 4

이것에 기반해서 계산된 모든 쌍의 USADO:
- 1번 ↔ 3번: min(3,2) = **2**
- 1번 ↔ 4번: min(3,4) = **3**  
- 3번 ↔ 4번: min(2,4) = **2**

### 질문별 해설

1. **K=1, 동영상 2번**: 
   - 2→1 (USADO 3 ≥ 1) ✓
   - 2→3 (USADO 2 ≥ 1) ✓
   - 2→4 (USADO 4 ≥ 1) ✓
   - **답: 3개**

2. **K=4, 동영상 1번**: 
   - 1→2 (USADO 3 < 4) ✗
   - 1→3 (USADO 2 < 4) ✗
   - 1→4 (USADO 3 < 4) ✗
   - **답: 0개**

3. **K=3, 동영상 1번**: 
   - 1→2 (USADO 3 ≥ 3) ✓
   - 1→3 (USADO 2 < 3) ✗
   - 1→4 (USADO 3 ≥ 3) ✓
   - **답: 2개**
